8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.




```python
x_tmp = x.copy()
#a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.
for column in x_tmp:
    most_common_value = x_tmp[column].value_counts().idxmax()
    print(f"Most common value in column {column}: {most_common_value}")

columns_to_fillna = ['EthnicGroup', 'ParentMaritalStatus', 'WklyStudyHours']# columns to fill

for column in columns_to_fillna:
    most_common_value = x_tmp[column].value_counts().idxmax()
    x_tmp[column].fillna(most_common_value, inplace=True)
#x_tmp.isna().sum() =>we add this to code to see if values are filled now


#b) For the feature ParentEduc replace NaN value with a string *no formal education*.
x_tmp["ParentEduc"].fillna("no formal educaiton", inplace=True)
#x_tmp.isna().sum() =>we add this to code to see if values are filled now


#c) For the feature TestPrep replace NaN value with a string *none*.
x_tmp["TestPrep"].fillna("none", inplace=True)
#x_tmp.isna().sum() =>we add this to code to see if values are filled now


#d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 
x_tmp.dropna(subset=['PracticeSport', 'IsFirstChild'], inplace=True)
x_tmp.reset_index(drop=True, inplace=True)

#e) For the feature NrSiblings replace the NaN value with the median value.
mean_value=x_tmp.NrSiblings.median()
x_tmp["NrSiblings"].fillna(mean_value, inplace=True)

#f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.
x_tmp["TransportMeans"].fillna("walking", inplace=True)



```

9. Are there inconsistencies in the IsFirstChild column? Replace the inconsistent entries so they match the rest of the values. 

```python
unique_values = x_tmp['IsFirstChild'].unique()
print("Unique values in IsFirstChild column:", unique_values) #to see what are all values in this column

inconsistent_entries_no = ['nein', 'nope', 'noo']
correct_value_no='no'
inconsistent_entries_yes = ['Yes','ye','si','yeah','ys','yeah boy','es']
correct_value_yes='yes'
x_tmp['IsFirstChild'].replace(inconsistent_entries_no, correct_value_no, inplace=True)
x_tmp['IsFirstChild'].replace(inconsistent_entries_yes, correct_value_yes, inplace=True)

```

10. The final step to preparing the dataset for classification is transforming the non-numeric variables to numeric variables. 
For features Gender and LunchType use OneHot Encoder, and for EthnicGroup, ParentEduc, TestPrep, ParentMaritalStatus, PracticeSport, IsFirstChild, TransportMeans and WklyStudyHours use Label Encoder. 

```python
x_tmp3=x_tmp
columns_to_labelencode=['EthnicGroup', 'ParentEduc', 'TestPrep', 'ParentMaritalStatus','PracticeSport', 'IsFirstChild', 'TransportMeans', 'WklyStudyHours']
label_encoder = LabelEncoder()
for column in columns_to_labelencode:
    x_tmp3[column] = label_encoder.fit_transform(x_tmp3[column])
print(x_tmp3.info())

#columns_to_onehotencode=['Gender', 'LunchType']
#X_tmp3.to_csv('students_cleaned.csv')
encoder = OneHotEncoder(handle_unknown="ignore")
encoder.fit(x_tmp3[['Gender']])
# we transform the data
X_tmp3= encoder.transform(x_tmp3[['Gender']])
X_tmp3= pd.DataFrame(X_tmp3.toarray(), columns=encoder.get_feature_names(['Gender']))
X_tmp3
Xfinal = pd.concat((x_tmp3, X_tmp3), axis=1)
Xfinal


```

```python
encoder.fit(Xfinal[['LunchType']])
#we transform the data
X_tmp4= encoder.transform(x_tmp3[['LunchType']])
X_tmp4= pd.DataFrame(X_tmp4.toarray(), columns=encoder.get_feature_names(['LunchType']))
X_tmp4
Xfinal2 = pd.concat((Xfinal, X_tmp4), axis=1)
Xfinal2
Xfinal2.to_csv('students_cleaned.csv')

```
